```python
import matplotlib.pyplot as plt

# Assuming the means and standard errors of the means (sem) are provided as follows:
a = [-274.32906433105467, -264.3991537475586]  # iteration_0_prompt_95 mean, iteration_1_prompt_95_mean
b = [-423.55554916381834, -419.1177679443359]  # iteration_0_prompt_85 mean, iteration_1_prompt_85_mean
c = [4.043953682990372, 4.034743665417116]    # iteration_0_prompt_95 sem, iteration_1_prompt_95 sem
d = [30.076707539972528, 29.753612903346202]    # iteration_0_prompt_85 sem, iteration_1_prompt_85 sem

# Define the iterations as x-axis labels
iterations = ['Iteration 0', 'Iteration 1']
iterations_numeric = [0, 1]  # Numeric x-axis locations for the iterations
marker_size = 6

# Plotting
plt.figure(figsize=(5, 6))
ax = plt.axes()
ax.set_facecolor("whitesmoke")

plt.xticks(range(2), iterations)  # Set custom labels for the ticks
plt.xlim(-0.5, 1.5)  # Set the limit so the x-axis will start a bit before 0 and end a bit after 1

# Error bars
ax.errorbar(iterations_numeric, a, yerr=c, fmt='o', color='#e86464', label='Prompt-95', markersize=marker_size)
ax.errorbar(iterations_numeric, b, yerr=d, fmt='o', color='#4285f4', label='Prompt-85', markersize=marker_size)

# Adding dotted lines
plt.plot(iterations_numeric, a, linestyle='dotted', color='#e86464')
plt.plot(iterations_numeric, b, linestyle='dotted', color='#4285f4')

# Title and labels
ax.set_title('Aside: Why do shuffled results improve at all?\nAverage Log-Probability of Shuffled Desired Results')
ax.set_xlabel('Iterations')
ax.set_ylabel('Log-Probability')
# ax.set_ylim([-475, -200])



# Show plot
plt.grid(True, color='white', linestyle='-', linewidth=0.9)
# Legend
ax.legend()
plt.tight_layout()
plt.savefig('log-prob.png')
plt.show()

```
